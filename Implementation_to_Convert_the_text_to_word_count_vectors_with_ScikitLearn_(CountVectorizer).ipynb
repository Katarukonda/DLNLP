```python
from sklearn.feature_extraction.text import CountVectorizer

# Example documents
documents = [
    "Machine learning is fascinating and challenging.",
    "Natural language processing is a key component of AI applications.",
    "Deep learning models require large amounts of data.",
    "Supervised and unsupervised learning are fundamental concepts in ML."
]

# Create an instance of CountVectorizer
vectorizer = CountVectorizer()

# Fit the vectorizer to the documents and transform the data
X = vectorizer.fit_transform(documents)

# Get the feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Print the feature names
print("Feature names:")
print(feature_names)

# Print the transformed sparse matrix (word count vectors)
print("\nTransformed matrix (word count vectors):")
print(X.toarray())
```
